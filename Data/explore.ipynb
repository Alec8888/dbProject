# Example Data: Temperature

```python
import pandas as pd
```

```python
data = pd.read_csv('../../dbProjectData/city_temperature.csv')
data.head()
```

```python
dataStats = data['AvgTemperature'].describe()
dataStats
```

There are 2,906,327 records in the dataset. This exceed the 250,000 requirement.

```python
unique_regions = data['Region'].unique()
unique_countries = {region: data[data['Region'] == region]['Country'].unique() for region in unique_regions}

unique_regions, unique_countries
```

```python
import matplotlib.pyplot as plt

us_data = data[data['Country'] == 'US']
yearly_avg_temp = us_data.groupby('Year')['AvgTemperature'].mean()

# Plot
plt.figure(figsize=(14, 6))
yearly_avg_temp.plot()
plt.title('Average Temperature Over Years in the US')
plt.xlabel('Year')
plt.ylabel('Average Temperature (F)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
